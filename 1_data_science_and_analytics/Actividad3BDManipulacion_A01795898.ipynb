Recuerda que:


*   Una **clave primaria** es un atributo (o conjunto) que identifica unívocamente a cada registro en la tabla.
*   Una **clave foránea** (externa o ajena) es un atributos (o conjunto) en una tabla que es una clave primaria en otra (o posiblemente la misma) tabla.
*   Las **relaciones** son las líneas que conectan una tabla con otra y el extremo determina la cardinalidad. Las relaciones con línea continua (identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser foránea y primaria (al mismo tiempo) de otra. Las relaciones con línea discontinua (no identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser sólo foránea en otra.

```python
# Aqui voy a crear una función que a partir de una string (la query) devuelva
# el data frame, para evitar ser tan redundante en cada paso

def execute_query(text):
    
    global conn
    query = sqla.text(text)
    df = pd.read_sql(query, conn)
    
    return df
```

```python
print('Empleados ordenados por nombre.')

# Nuevamente asumamos que se quiere el nombre de los empleados y no toda la información:
sorted_employees = employees[['firstName', 'lastName']].sort_values(by='firstName')
sorted_employees.reset_index(drop=True, inplace=True)
sorted_employees
```

```python
# TODA la información de los empleados ordenada por firstName

aux = employees[['firstName', 'lastName', 'employeeNumber', 'extension', 'jobTitle', 'officeCode', 'reportsTo', 'email']]
aux = aux.sort_values(by='firstName')
aux.reset_index(drop=True, inplace=True)
aux
```

```python
print('Nombre y teléfono de los clientes de Nueva York (NYC)')

aux = customers[customers['city']=='NYC']
aux = aux[['customerName', 'phone']]
aux.reset_index(drop=True, inplace=True)
aux
```

```python
print('Código y nombre de los productos del vendedor “Gearbox Collectibles” que tengan menos de 1000 unidades en stock.')

aux = products[products['productVendor']=='Gearbox Collectibles']
aux = aux[aux['quantityInStock']<1000][['productCode', 'productName']]
aux.reset_index(drop=True, inplace=True)
aux
```

```python
print('Tres productos más caros, desde el punto de vista de los comercializadores (buyPrice).')

aux = products.sort_values(by='buyPrice', ascending=False)[['productName', 'buyPrice']].iloc[0:3]
aux.reset_index(drop=True, inplace=True)
aux
```

```python
print('Cantidad de productos por línea de producto (no las existencias en inventario).')
products.groupby(by='productLine')['productCode'].nunique().reset_index(name='cantidadDeProductos')
```

```python
print('Cantidad de empleados por país (tomando en cuenta la ubicación de la oficina')

aux = pd.merge(employees, offices, how='inner', on='officeCode')

aux.groupby(by='country')['employeeNumber'].nunique().reset_index(name='NumTotalEmployees')
```

```python
print('Promedio de los pagos de los clientes de España (sin incluir aquellos que no poseen ningún pago)')

ids = list(customers[customers['country']=='Spain']['customerNumber'].unique())
aux = payments[payments['customerNumber'].isin(ids)]

aux.groupby(by='customerNumber')['amount'].mean().reset_index(name='averageAmount')
```

12.	Investiga cómo leer la colección `PET_OWNER` y mostrar su contenido en un dataframe. Asegúrate de incluir el id en el resultado

```python
# El sigeuinte paso es para obtener los documentos dentro de dicha colección
docs = db.collection('PET_OWNER').stream()

# Dada la inspección visual del documentos vemos que cada documento tiene información relevante
# La cual podemos parsear y darle un formato tabular

# Defino mis listas auxiliar que después se iran llenando en cada iteración
data = []
ids = []

#Revise que hay un documento más que el de la rubrica, por ende lo voy a filtrar
# Ya que tienen Nan en todas sus columnas de información

for doc in docs:
    info = doc.to_dict()
    if 'ownerFirstName' in info: 
        ids.append(doc.id)
        data.append(info)

# Finalmente creo data frame indexado por los id
df = pd.DataFrame(data, index=ids)
# Renombro el indice y reordeno columna para tener coincidencia exacta con el que pide la rubrica
df = df.rename_axis('id')
df = df[['ownerLastName', 'email', 'phone', 'ownerFirstName']]

df
```
